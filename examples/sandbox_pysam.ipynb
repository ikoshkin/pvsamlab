```python
import requests

# Define API URL
api_url = "https://developer.nrel.gov/api/nsrdb/v2/solar/psm3-download.csv"

# Define parameters
params = {
    "api_key": "42919c86-c4bb-4447-9e42-515672ca3bbd",
    "lat": 32.7767,
    "lon": -96.7970,
    "names": "2023",  # Year (Historical Data)
    "interval": "60",  # Time resolution (e.g., 60 minutes)
    "utc": "false"  # Convert to local time
}

# Fetch weather data
response = requests.get(api_url, params=params)

# Save if valid
if response.status_code == 200:
    with open("weather_psm3_2023.csv", "w") as f:
        f.write(response.text)
    print("✅ Weather data saved: weather_psm3_2023.csv")
else:
    print(f"❌ Error fetching weather data: {response.status_code}")
```
